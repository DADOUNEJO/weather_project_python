## Exploratory Analysis and Visualization

Here we will be using matplotlib and seaborn libraries which provides necessary functions for data visualisation.


```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
df.describe()
```

```python

plt.bar(df["Temperature (C)"][:30], df["Wind Speed (km/h)"][:30], color ='maroon',
        width = 0.1)
plt.title('Plot of Visibility and Dew Point')
plt.xlabel('Visibility')
plt.xticks(rotation=75)
plt.ylabel('Dew Point')
```

Here we can see that if dew point is less then visibility is more.
